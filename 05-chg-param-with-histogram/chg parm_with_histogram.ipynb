```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
BINS = 6
#############################################
#                                           #
#            Loading data file              #
#                                           #
#############################################
data = np.loadtxt(
                  #'/home/flgomez10/lum-mass/data/MD_3840_Planck1/BDM/small_cell_cat.dat',
                  #'/home/flgomez10/lum-mass/data/MD_3840_Planck1/BDM/test.dat',
                  '/home/flgomez10/lum-mass/data/MD_3840_Planck1/BDM/reduced_catshortV_007.dat',
                  # Reduce the memory usage from 'float64'(default) to 'float32'
                  #'float32',
                  usecols=(0,1,2,3), skiprows=17)

#Observational data
Obs_data = np.array([#  Mag,  Lum   , up_err, low_err #
                     [-22.5, 2.66E-8, 9.08E-9, 7.78E-8],
                     [-22.0, 2.18E-6, 8.70E-7, 9.70E-6],
                     [-21.5, 1.45E-5, 2.88E-6, 2.92E-5],
                     [-21.0, 1.29E-4, 7.06E-5, 2.19E-4],
                     [-20.5, 2.30E-4, 9.34E-5, 5.77E-4]
                    ])
Obs_data_log = array([
                        [ -22.5, -7.57511836, -8.04191415, -7.1090204 ],
                        [ -22.0, -5.66154351, -6.06048075, -5.01322827],
                        [ -21.5, -4.838632  , -5.54060751, -4.53461715],
                        [ -21.0, -3.88941029, -4.1511953 , -3.65955589],
                        [ -20.5, -3.63827216, -4.02965312, -3.23882419]
                      ])
```

```python
#############################################
#                                           #
#            Halo Mass Function             #
#                                           #
#############################################
M_halos_list = np.log10(data[:,3])
Mass_bin_width = (12.5-10.5)/5
Cell_Lenght = 1000 #Mpc h-1
HIST = histogram(M_halos_list ,bins=linspace(10.5,13.5,7) )[0]
HIST = HIST/( Cell_Lenght**3 * Mass_bin_width)
```

```python
# Plotting Mass Density Function
semilogy(linspace(10.5,13.0,6), HIST,"r-o")
#axis([12.5,-20,-8,-2])
plt.title("Mass Density Function")
plt.xlabel(r"$\log_{10} M_\odot$")
plt.ylabel(r"$\log \frac{dM}{dN}\left( Mpc^3 h^{-1} \right)$")
axis([10,13,1.0e-8,1.0e-1])
#plt.show()
```

```python
#############################################
#                                           #
#            Luminosity Function            #
#                                           #
#############################################
Magnitude_bin_width = 0.5

alpha = 10**(2.91)
beta  =0.649

# Best Fitting Parameters:
# beta = 0.649 +/- 0.040
# log10(alpha) = 2.91 +/- 0.28

#0.649 812.830516164 0.786633438801
# 0.71 141.253754462 0.652875287612
# 0.605 2818.38293126 0.750125985791
# 0.545 15135.6124844 0.564951883804
# 0.52 31622.7766017 0.440764488265
# 0.5 56234.132519
# 0.488888888889 100000.0 5.39872344137
# 0.6 4641.58883361 3.76338143897
# $L = \alpha * Halo_Mass ^ { \beta}$

Log_Luminosity_halo_list = np.log10(alpha) + beta*M_halos_list 
Magnitude_UV_galaxy_list = 5.61 - 2.5 *  Log_Luminosity_halo_list[:] 

HIST = histogram(Magnitude_UV_galaxy_list,bins=(-22.5,-22.0,-21.5,-21.0,-20.5,-20.0))[0]
HIST = HIST/(Magnitude_bin_width*Cell_Lenght**3)
HIST = log10(HIST)

Sum_Err = 0.0
for i in range(5):
    if (HIST[i] > -10) & (HIST[i]<0):
        Sum_Err += (HIST[i] - Obs_data_log[i,1])**2
    else:
        Sum_Err +=20
print beta, alpha, exp(-Sum_Err)


```
